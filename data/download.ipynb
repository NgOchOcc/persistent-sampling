```python
from datasets import load_dataset

math500 = load_dataset('HuggingFaceH4/MATH-500')
gsm8k = load_dataset('openai/gsm8k', 'main')
```

```python
math500_test = [item for item in math500['test']]
gsm8k_test = [{
    'problem': item['question'],
    'answer': item['answer']} for item in gsm8k['test']]
```

```python
import json

json.dump(math500_test, open('math500.json', 'w'), indent=4)
json.dump(gsm8k_test, open('gsm8k.json', 'w'), indent=4)
```
